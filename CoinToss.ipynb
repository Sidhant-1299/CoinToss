```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class Coin:
    #Heads = 1
    #Tails = 0
    def toss_ten_times(self):
        return np.random.randint(0,2,size=10)
```

```python
def test_coin_flip(n):
    #Ten is total flips
    #We want the times heads appear :ie 1
    c = Coin()
    return [c.toss().sum()/10 for i in range(n)]
```

```python
N = 1000000
```

```python
plt.figure(figsize=(10,6))
heads_ratio = test_coin_flip(N)
plt.title(f"Probability distribution of landing heads on {N} flips")
sns.histplot(data = heads_ratio,bins=11)
```
